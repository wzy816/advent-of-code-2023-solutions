```python
with open("./input.txt", "r") as f:
    weight = [[int(i) for i in l] for l in f.read().split("\n")]

rows = len(weight)
cols = len(weight[0])

# node = (i,j,next_move) left,right,up,down
nodes = []
for i in range(rows):
    for j in range(cols):
        if i > 0:
            nodes.append((i, j, "u"))
        if i < rows - 1:
            nodes.append((i, j, "d"))
        if j > 0:
            nodes.append((i, j, "l"))
        if j < cols - 1:
            nodes.append((i, j, "r"))

# graph = {node: edges}
# edges = [(next_node,edge weight), ()]
graph = {}
for node in nodes:
    graph[node] = []

# build edges
for i in range(rows):
    for j in range(cols):

        edges = []

        for dist in range(1, 4):
            edges.append([(i, j, "u"), (i - dist, j, "l")])
            edges.append([(i, j, "u"), (i - dist, j, "r")])
            edges.append([(i, j, "d"), (i + dist, j, "l")])
            edges.append([(i, j, "d"), (i + dist, j, "r")])

            edges.append([(i, j, "l"), (i, j - dist, "u")])
            edges.append([(i, j, "l"), (i, j - dist, "d")])
            edges.append([(i, j, "r"), (i, j + dist, "u")])
            edges.append([(i, j, "r"), (i, j + dist, "d")])

        for (
            cur_node,
            next_node,
        ) in edges:
            if cur_node not in graph:
                continue
            if next_node not in graph:
                continue
            next_i, next_j, _ = next_node
            if (
                next_node[0] < 0
                or next_node[0] > rows - 1
                or next_node[1] < 0
                or next_node[1] > cols - 1
            ):
                continue
            # cal weight
            w = 0
            if cur_node[2] == "u":
                for i_ in range(cur_node[0] - 1, next_node[0] - 1, -1):
                    w += weight[i_][next_node[1]]
            if cur_node[2] == "d":
                for i_ in range(cur_node[0] + 1, next_node[0] + 1):
                    w += weight[i_][next_node[1]]
            if cur_node[2] == "l":
                for j_ in range(cur_node[1] - 1, next_node[1] - 1, -1):
                    w += weight[next_node[0]][j_]
            if cur_node[2] == "r":
                for j_ in range(cur_node[1] + 1, next_node[1] + 1):
                    w += weight[next_node[0]][j_]
            graph[cur_node].append((next_node, w))

# dijstra
heat_lost = {}
for node in nodes:
    heat_lost[node] = float("inf")

heat_lost[(0, 0, "r")] = weight[0][0]
heat_lost[(0, 0, "d")] = weight[0][0]

q = [(0, 0, "r"), (0, 0, "d")]

while q:
    q_ = []
    for node in q:

        edges = graph[node]
        for next_node, w in edges:
            if heat_lost[node] + w < heat_lost[next_node]:
                heat_lost[next_node] = heat_lost[node] + w
                # only visit again if new w
                q_.append(next_node)
    q = q_[:]

print(
    min(
        [
            heat_lost[(rows - 1, cols - 1, "l")] - weight[0][0],
            heat_lost[(rows - 1, cols - 1, "u")] - weight[0][0],
        ]
    )
)
```
