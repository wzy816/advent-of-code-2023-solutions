```python
with open("./input.txt", "r") as f:
    data = f.read().splitlines()

tiles = data[:]


rows = len(tiles)
cols = len(tiles[0])
ans = 0
for i in range(rows):
    for j in range(cols):
        if tiles[i][j] == "S":

            d = []
            d.append([i - 1, j, "south", 1])
            d.append([i + 1, j, "north", 1])
            d.append([i, j - 1, "east", 1])
            d.append([i, j + 1, "west", 1])

            while len(d) > 0:
                _i, _j, _last, step = d.pop()

                if _i < 0 or _i >= rows or _j < 0 or _j >= cols:
                    continue
                if tiles[_i][_j] == ".":
                    continue
                if tiles[_i][_j] == "S":
                    ans = step
                    break
                if tiles[_i][_j] == "-":
                    if _last == "east":
                        d.append([_i, _j - 1, "east", step + 1])
                    if _last == "west":
                        d.append([_i, _j + 1, "west", step + 1])
                if tiles[_i][_j] == "|":
                    if _last == "south":
                        d.append([_i - 1, _j, "south", step + 1])
                    if _last == "north":
                        d.append([_i + 1, _j, "north", step + 1])
                if tiles[_i][_j] == "L":
                    if _last == "east":
                        d.append([_i - 1, _j, "south", step + 1])
                    if _last == "north":
                        d.append([_i, _j + 1, "west", step + 1])
                if tiles[_i][_j] == "J":
                    if _last == "west":
                        d.append([_i - 1, _j, "south", step + 1])
                    if _last == "north":
                        d.append([_i, _j - 1, "east", step + 1])
                if tiles[_i][_j] == "7":
                    if _last == "west":
                        d.append([_i + 1, _j, "north", step + 1])
                    if _last == "south":
                        d.append([_i, _j - 1, "east", step + 1])
                if tiles[_i][_j] == "F":
                    if _last == "east":
                        d.append([_i + 1, _j, "north", step + 1])
                    if _last == "south":
                        d.append([_i, _j + 1, "west", step + 1])
print(int(ans / 2))
```

```python
with open("./input.txt", "r") as f:
    data = f.read().splitlines()

    size = len(data[0])

    # padding with . if S is on border
    tiles = [["."] * (size + 2)]
    for row in data:
        tiles.append(["."] + list(row) + ["."])
    tiles += [["."] * (size + 2)]


rows = len(tiles)
cols = len(tiles[0])

for i in range(rows):
    for j in range(cols):
        if tiles[i][j] == "S":
            d = []
            d.append([i - 1, j, "south"])
            d.append([i + 1, j, "north"])
            d.append([i, j - 1, "east"])
            d.append([i, j + 1, "west"])

            while len(d) > 0:
                _i, _j, _last = d.pop()

                if _i < 0 or _i >= rows or _j < 0 or _j >= cols:
                    continue

                if tiles[_i][_j] == "." or tiles[_i][_j] == "X" or tiles[_i][_j] == "Y":
                    continue

                if tiles[_i][_j] == "S":
                    # eye check input.txt, S is actually F, thus Y
                    tiles[_i][_j] = "Y"
                    break

                # mark - 7 F as Y, not wall
                # mark | L J as X, wall (lower bottom bar)
                if tiles[_i][_j] == "-":
                    if _last == "east":
                        d.append([_i, _j - 1, "east"])
                        tiles[_i][_j] = "Y"
                    if _last == "west":
                        d.append([_i, _j + 1, "west"])
                        tiles[_i][_j] = "Y"
                if tiles[_i][_j] == "|":
                    if _last == "south":
                        d.append([_i - 1, _j, "south"])
                        tiles[_i][_j] = "X"
                    if _last == "north":
                        d.append([_i + 1, _j, "north"])
                        tiles[_i][_j] = "X"
                if tiles[_i][_j] == "L":
                    if _last == "east":
                        d.append([_i - 1, _j, "south"])
                        tiles[_i][_j] = "X"
                    if _last == "north":
                        d.append([_i, _j + 1, "west"])
                        tiles[_i][_j] = "X"
                if tiles[_i][_j] == "J":
                    if _last == "west":
                        d.append([_i - 1, _j, "south"])
                        tiles[_i][_j] = "X"
                    if _last == "north":
                        d.append([_i, _j - 1, "east"])
                        tiles[_i][_j] = "X"
                if tiles[_i][_j] == "7":
                    if _last == "west":
                        d.append([_i + 1, _j, "north"])
                        tiles[_i][_j] = "Y"
                    if _last == "south":
                        d.append([_i, _j - 1, "east"])
                        tiles[_i][_j] = "Y"
                if tiles[_i][_j] == "F":
                    if _last == "east":
                        d.append([_i + 1, _j, "north"])
                        tiles[_i][_j] = "Y"
                    if _last == "south":
                        d.append([_i, _j + 1, "west"])
                        tiles[_i][_j] = "Y"

# ray casting algorithm to check point in polygon
ans = 0
for i in range(rows):
    cnt = 0
    for j in range(cols):
        tile = tiles[i][j]
        if tile == "X":
            cnt += 1
        elif tile != "Y":
            if cnt % 2 == 1 and i != 0 and i != rows - 1 and j != 0 and j != cols - 1:
                ans += 1

print(ans)
```
