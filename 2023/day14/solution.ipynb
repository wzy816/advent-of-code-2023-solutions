```python
def transpose(grid):
    return [[l[i] for l in grid] for i in range(len(grid[0]))]


def flip_h(grid):
    return [l[::-1] for l in grid]


def flip_v(grid):
    return transpose(flip_h(transpose(grid)))


def flatten(grid):
    return "".join(["".join(l) for l in grid])


def copy(grid):
    return [l[:] for l in grid]


def cal_weight(grid):
    ans = 0
    max_weight = len(grid)
    for i, row in enumerate(grid):
        ans += row.count("O") * (max_weight - i)
    return ans


def roll_north(grid):
    cols = len(grid[0])

    new_columns = []

    for c in range(cols):

        column = [l[c] for l in grid]

        round = []
        space = []

        new_column = []
        for r in column:
            if r == "O":
                round.append(r)
            elif r == ".":
                space.append(r)
            elif r == "#":
                if round:
                    new_column += round
                if space:
                    new_column += space
                new_column.append(r)
                round = []
                space = []
        if round:
            new_column += round
        if space:
            new_column += space

        new_columns.append(new_column)

    return transpose(new_columns)


def roll_south(grid):
    return flip_v(roll_north(flip_v(grid)))


def roll_east(grid):
    return transpose(roll_south(transpose(grid)))


def roll_west(grid):
    return flip_h(roll_east(flip_h(grid)))


with open("./input.txt", "r") as f:
    data = f.read().split("\n")
    grid = [[l for l in line] for line in data]

    d_after_i = {}
    g1 = copy(grid)

    cycles = 1000000000 - 1
    c = 0
    # find loop point
    while c < cycles:
        g1 = roll_north(g1)
        g1 = roll_west(g1)
        g1 = roll_south(g1)
        g1 = roll_east(g1)
        f = flatten(g1)
        if f not in d_after_i:
            d_after_i[f] = (c, cal_weight(g1))
        else:
            c2 = d_after_i[f][0]
            break

        c += 1

    print(c, c2)
    # 0 => 111 => 112(84) => 139(111) => ...
    final_c = (cycles - c2) % (c - c2) + c2

    for k, v in d_after_i.items():
        if v[0] == final_c:
            print(v[1])
            break
```
