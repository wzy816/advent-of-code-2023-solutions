```python
with open("./input.txt", "r") as f:
    data = f.read().splitlines()

from collections import Counter


def is_five(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 1


def is_four(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 2 and 4 in c.values()


def is_fullhouse(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 2 and 2 in c.values()


def is_three(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 3 and 3 in c.values()


def is_two_pair(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 3 and 2 in c.values()


def is_one_pair(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 4


def is_high(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 5


def rank(hands):
    if is_five(hands):
        return 6
    elif is_four(hands):
        return 5
    elif is_fullhouse(hands):
        return 4
    elif is_three(hands):
        return 3
    elif is_two_pair(hands):
        return 2
    elif is_one_pair(hands):
        return 1
    elif is_high(hands):
        return 0
    return 0


def is_lt_of_same_rank(hands1, hands2):
    p = {
        l: -i
        for i, l in enumerate(
            ["A", "K", "Q", "J", "T", "9", "8", "7", "6", "5", "4", "3", "2"]
        )
    }

    for l1, l2 in zip(list(hands1), list(hands2)):
        if p[l1] < p[l2]:
            return True
        elif p[l1] > p[l2]:
            return False


class Hand:
    def __init__(self, hands, bid):
        self.hands = hands
        self.bid = bid

    def __lt__(self, other):
        r1 = rank(self.hands)
        r2 = rank(other.hands)
        if r1 == r2:
            return is_lt_of_same_rank(self.hands, other.hands)
        return r1 < r2

    def __repr__(self) -> str:
        return f"{self.hands} {self.bid}"


cards = []
for line in data:
    hands, bid = line.split(" ")
    bid = int(bid)
    cards.append(Hand(hands, bid))

ans = 0
for i, card in enumerate(sorted(cards)):
    ans += card.bid * (i + 1)
print(ans)
```

```python
with open("./input.txt", "r") as f:
    data = f.read().splitlines()

from collections import Counter


def is_five(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 1


def is_four(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 2 and 4 in c.values()


def is_fullhouse(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 2 and 2 in c.values()


def is_three(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 3 and 3 in c.values()


def is_two_pair(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 3 and 2 in c.values()


def is_one_pair(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 4


def is_high(hands):
    c = Counter(list(hands))
    return len(c.keys()) == 5


def rank_with_j(hands):
    ans = rank(hands)
    if "J" not in hands:
        return ans

    for w in ["A", "K", "Q", "T", "9", "8", "7", "6", "5", "4", "3", "2"]:
        h = hands.replace("J", w)
        ans = max(ans, rank(h))
    return ans


def rank(hands):
    if is_five(hands):
        return 6
    elif is_four(hands):
        return 5
    elif is_fullhouse(hands):
        return 4
    elif is_three(hands):
        return 3
    elif is_two_pair(hands):
        return 2
    elif is_one_pair(hands):
        return 1
    elif is_high(hands):
        return 0
    return 0


def is_lt_of_same_rank(hands1, hands2):
    p = {
        l: -i
        for i, l in enumerate(
            ["A", "K", "Q", "T", "9", "8", "7", "6", "5", "4", "3", "2", "J"]
        )
    }

    for l1, l2 in zip(list(hands1), list(hands2)):
        if p[l1] < p[l2]:
            return True
        elif p[l1] > p[l2]:
            return False


class Hand:
    def __init__(self, hands, bid):
        self.hands = hands
        self.bid = bid

    def __lt__(self, other):
        r1 = rank_with_j(self.hands)
        r2 = rank_with_j(other.hands)
        if r1 == r2:
            return is_lt_of_same_rank(self.hands, other.hands)
        return r1 < r2

    def __repr__(self) -> str:
        return f"{self.hands} {self.bid}"


cards = []
for line in data:
    hands, bid = line.split(" ")
    bid = int(bid)
    cards.append(Hand(hands, bid))

ans = 0
for i, card in enumerate(sorted(cards)):
    ans += card.bid * (i + 1)
print(ans)
```
